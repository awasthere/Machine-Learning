##Manipulating Data

```python
data["Make"]=data["Make"].str.upper()
```

```python
data
```

```python
#column from series
seats_col=pd.Series([5,5,5,5,5])
#new coloumn called seats
data["Seats"]=seats_col
data
```

```python
data["Seats"].fillna(5,inplace=True)
data
```

```python
fuel=[4,6,2,6,8,4,5,6,8,9]
data["Fuel per 100km"]=fuel
data["Total fuel used"]=data["Odometer (KM)"]/100*data["Fuel per 100km"]
data
```

```python
data.to_csv("new-car-sales.csv")
```
