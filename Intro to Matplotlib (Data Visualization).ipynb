## Two Options for subplots

```python
# method 1
fig, ((ax1,ax2), (ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax1.plot(x,x/2);
ax2.scatter(np.random.random(10),np.random.random(10));
ax3.bar(butter.keys(),butter.values());
ax4.hist(np.random.randn(1000));
```

```python
#subplots option2
fig,ax=plt.subplots(nrows=2,ncols=2,figsize=(10,5))
#plot each different index
ax[0,0].plot(x,x/2);
ax[0,1].bar(butter.keys(),butter.values());
ax[1,0].scatter(np.random.random(10),np.random.random(10));
ax[1,1].hist(np.random.randn(1000));
```

## Which one should you use? (pyplot VS matplotlib OO method)

1. When plotting something quickly, okay to use plot method
2. When plotting something advanced, use the OO method

```python
heart_dis
```

```python
over_50=heart_dis[heart_dis['age']>50]
len(over_50)
```

```python
over_50.plot(x='age',y='chol',kind='scatter',c='target');
```

```python
fig,ax=plt.subplots(figsize=(8,6))
over_50.plot(kind='scatter',x='age',y='chol',c='target',ax=ax);
```

```python
#setting up plot
fig,ax=plt.subplots(figsize=(7,5))
#plotting the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);
#styling the plot
ax.set(title="Heart disease",xlabel="Age",ylabel="Cholestrol");
#adding the legend
ax.legend(*scatter.legend_elements(),title='target');
#adding horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
over_50
```

```python
#subplot 1
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(11,10),sharex=True)
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target']);
#styling the plot
ax0.set(title="Heart disease",ylabel="Cholestrol");
#adding the legend
ax0.legend(*scatter.legend_elements(),title='target');
#adding horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle='--');

#subplot 2
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'])
#styling the plot
ax1.set(title="Heart disease",xlabel="Age",ylabel="thalach")
#adding the legend
ax1.legend(*scatter.legend_elements(),title='target')
#adding horizontal line
ax1.axhline(over_50['thalach'].mean(),linestyle='--')
fig.suptitle("Heart disease analysis",fontsize=16,fontweight='bold');
```

```python
import matplotlib.pyplot as plt
std=["5th","6th","7th","8th","9th","10th","11th","12th"]
perc=[85,72,70,73,60,80,86,85]
fig,ax=plt.subplots()
ax.plot(std,perc);
ax.set(title="Aryan's performance",xlabel="class",ylabel="percentage (%)");
```

```python
#setting up plot
fig,ax=plt.subplots(figsize=(7,5))
#plotting the data
scatter=ax.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="winter");
#styling the plot
ax.set(title="Heart disease",xlabel="Age",ylabel="Cholestrol");
#adding the legend
ax.legend(*scatter.legend_elements(),title='target');
#adding horizontal line
ax.axhline(over_50['chol'].mean(),linestyle='--');
```

```python
#subplot 1
fig,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(9,7),sharex=True)
scatter=ax0.scatter(x=over_50['age'],y=over_50['chol'],c=over_50['target'],cmap="winter")
#styling the plot
ax0.set(title="Heart disease and Cholesterol level",ylabel="Cholesterol");
ax0.set_xlim([50,80])
#adding the legend
ax0.legend(*scatter.legend_elements(),title='target');
#adding horizontal line
ax0.axhline(over_50['chol'].mean(),linestyle='--');

#subplot 2
scatter=ax1.scatter(x=over_50['age'],y=over_50['thalach'],c=over_50['target'],cmap="winter")
#styling the plot
ax1.set(title="Heart disease and Max heart rate",xlabel="Age",ylabel="Max heart rate")
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
#adding the legend
ax1.legend(*scatter.legend_elements(),title='target')
#adding horizontal line
ax1.axhline(over_50['thalach'].mean(),linestyle='--')
fig.suptitle("Heart disease analysis",fontsize=16,fontweight='bold');
fig.savefig("heart-disease-analysis.jpg")
```
